```python
def f2(x,w):
    return (x-w)*x*(x+2)
x = np.linspace(-3,3,100)

plt.plot(x, f2(x,2), color="black", label="$w=2$")
plt.plot(x, f2(x,1), color="cornflowerblue", label="$w=1$")
plt.legend(loc="upper left") #凡例表示
plt.ylim(-15,15) #y軸の範囲
plt.title("$f_2(x)$")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.grid(True) #グリッド
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f3(x0, x1):
    ans = (2*x0**2 + x1**2) * np.exp(-(2*x0**2 + x1**2))
    return ans

xn = 9
x0 = np.linspace(-2, 2, xn)
x1 = np.linspace(-2, 2, xn)
y = np.zeros((len(x0), len(x1)))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])

print(x0)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D #3次元表示するためにmpl_toolkits.mplot3dのAxes3Dをimport: from ライブラリ名 import 関数名
#import mpl_toolkits.mplot3d #このようにライブラリを呼び出しておき、後からライブラリ名.関数名としなくても良いらしい。
#参照: モジュール（pythonのfile）、オブジェクト（dataやdataを扱う関数の集まり）、メソッド（関数そのもの）

def f3(x0, x1):
    ans = (2*x0**2 + x1**2) * np.exp(-(2*x0**2 + x1**2))
    return ans

xn = 9
x0 = np.linspace(-2, 2, xn)
x1 = np.linspace(-2, 2, xn)

y = np.zeros((len(x0), len(x1)))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])

xx0, xx1 = np.meshgrid(x0, x1) #格子点を用意: x0,x1,yだと3D plotができない←要素数が違うx0,x1はベクトル, yは行列

plt.figure(figsize=(5,3.5))
ax = plt.subplot(1,1,1,projection='3d') #projection='3d'で3D座標系を指定。axにグラフのidを示す返り値を保存
ax.plot_surface(xx0,xx1,y,rstride=1,cstride=1,alpha=0.3,color='blue',edgecolor='black') #rstride x軸について何個の要素ずつでlineを引くかを指定。cstrideはy軸について。alphaは面の透明度を指定するoptionで0~1の指定、1に近づくほど不透明。
ax.set_zticks((0,0.2)) #目盛りの調整
ax.view_init(75,-95) #3次元グラフの向きを調整。引数1: 上下の回転角度を表し、0だと真横から見たグラフ、90だと真上から見たグラフ。引数2: 横の回転角度
plt.show()
```

```python
xn = 50 #解像度になる。contourは解像度を高めないと正確に表示されない。
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)

y = np.zeros((xn, xn))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1,i0] = f3(x0[i0],x1[i1])
        
xx0, xx1 = np.meshgrid(x0,x1)

plt.figure(1, figsize=(4,4))
cont = plt.contour(xx0, xx1, y, 5, colors='black') #等高線プロットを作成。5は高さのレベルを5段階に設定。
cont.clabel(fmt='%.2f', fontsize=8) #各等高線に数値を入れる。
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
plt.show()
```
